```python
import pandas as pd
import numpy as np
df = pd.read_csv("drive/My Drive/Colab Notebooks/EXP3_1.csv")
df = df.drop('TID',axis=1)
df1 = df.iloc[:,-1:]
df2 = df.iloc[:,:-1]
dict2 = {}
unqlis2 = df1.columns.unique()
for i in df2.columns:
  list3 = []
  w_avg = []
  unqlis = df2[i].unique()
  print()
  print("Finding for ",i)
  for k in unqlis:
    val = df2[df2[i] == k][i].count()/df2.shape[0]
    list3.append(val)
    g_vh=0
    countlist = []
    for a in df1.iloc[:,0].unique():
      c1 = 0;
      for j in range(df2.shape[0]):
        if df2[i][j] == k:
          if df1.iloc[j,0] == a:
            c1 = c1 + 1
      countlist.append(c1)
    for c in countlist:
      g_vh = g_vh + (c/sum(countlist)) **2
    w_avg.append((1-g_vh))
    print("Gini index of ",k," = ",(1-g_vh))
  w_ans = 0
  for m in range(len(w_avg)):
    w_ans = w_ans + (w_avg[m] * list3[m])
  print("Weighted Average of ",i," = ",w_ans)
  dict2[i] = w_ans

temp = min(dict2.values())
res = [key for key in dict2 if dict2[key] == temp]
```

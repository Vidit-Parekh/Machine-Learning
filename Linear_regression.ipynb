```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

ds = pd.read_csv("file.csv")

X = ds['X']
Y = ds['Y']
```

```python
x_m = np.mean(X)
y_m = np.mean(Y)

hr = 0
dr = 0
```

```python
for i in range(len(ds)):
    hr += (X[i]-x_m)*(Y[i]-y_m)
    dr += (X[i]-x_m)**2
m = hr/dr
print(m)
```

```python
c = y_m - m*x_m
print(c)
```

```python
l=[]
for i in range(len(ds)):
    yp = m*X[i]+c
    l.append(yp)
print(l)
```

```python
yp1 = np.array(l)
print(yp1)
```

```python
plt.scatter(X,Y)
plt.plot(X,yp1)
plt.show()
```

```python
n = len(X)
m = 0
c = 0
L = 0.01
epochs = 100
for i in range(epochs):
    for j in range(len(ds)):
        yp = m * X[j] + c
        dm = (-2/n)* (X[j]*(Y[j]-yp))
        dc = (-2/n)* (Y[j]-yp)
        m -= L * dm
        c -= L * dc
print(m,c)
```

```python
yp = m*X+c
plt.scatter(X,Y)
plt.plot(X,yp)
plt.show()
```
